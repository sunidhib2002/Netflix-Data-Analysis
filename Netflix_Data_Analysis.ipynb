```python
newData = movies['listed_in'].value_counts()[:10]
# using value_counts() -> we can get the values in the column with maximum occurence, 
# this value_counts() -> counts the occurence then return that list in descending order
# from here we can pick any 5 or 10 best values 
```

```python
newData
```

```python
newData = newData.reset_index()
```

```python
newData
```

```python
newData.rename(columns = {'index' : 'Category'}, inplace = True)
```

```python
newData.rename(columns = {'listed_in': 'count'}, inplace = True)
```

```python
plt.figure(figsize = (10, 10))
plt.title("Mostly watched categories")
sns.barplot(x = newData['count'], y = newData['Category'])
```
